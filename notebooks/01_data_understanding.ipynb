## 1.4 Dataset overlap

```python
# check how many accounts overlap between the two datasets
ret_accts = set(retention['Customer Account Number'].dropna().unique())
bob_accts = set(bob['account_number'].dropna().unique())

common = ret_accts & bob_accts
only_ret = ret_accts - bob_accts
only_bob = bob_accts - ret_accts

print(f'Retention unique accounts: {len(ret_accts):,}')
print(f'BoB unique accounts: {len(bob_accts):,}')
print(f'Common: {len(common):,}')
print(f'Only in Retention: {len(only_ret):,}')
print(f'Only in BoB: {len(only_bob):,}')
print(f'Overlap rate: {len(common)/len(ret_accts)*100:.1f}%')
```

## 1.5 Key Observations

- **BoB**: ~198K active records, 24K unique accounts
- **Overlap**: ~26% of retention accounts appear in BoB
